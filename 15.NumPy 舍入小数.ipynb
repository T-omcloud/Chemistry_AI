# NumPy 舍入小数

### Truncation
* 删除小数，并返回最接近零的浮点数。使用 trunc() 和 fix() 函数。

```python
# 使用 trunc()：
import numpy as np

arr = np.trunc([-3.1666, 3.6667])
print(arr)

# 使用 fix()：
arr = np.fix([-3.1666, 3.6667])

print(arr)
```

### 四舍五入
* around() 函数会根据四舍五入的规则对数字进行舍入。
* 如果小数点后一位数字或小数位数后的数字大于或等于5，则前一位数字或小数位数加1，否则不做任何操作。例如。四舍五入到小数点后 1 位，3.16666 是 3.2。

```python
# 将 3.1666 四舍五入到小数点后两位：
arr = np.around(3.1666, 2)

print(arr)
```

### 向下取整
* floor() 函数将小数四舍五入为最接近的较小整数。

```python
arr = np.floor([-3.1666, 3.6667])
print(arr)
```

### 向上取整
* ceil() 函数将小数四舍五入为最接近的大整数。

```python
arr = np.ceil([-3.1666, 3.6667])
print(arr)
```
