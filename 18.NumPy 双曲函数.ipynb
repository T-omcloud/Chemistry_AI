# NumPy 双曲函数

### 双曲函数（Hyperbolic Functions）
* NumPy 提供了 ufuncs sinh()、cosh() 和 tanh()，它们以弧度为单位获取值，并生成相应的 sinh、cosh 和 tanh 值。

```python
import numpy as np
x = np.sinh(np.pi/2)
print(x)

# 求 arr 中所有值的 cosh 值：
arr = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])
x = np.cosh(arr)
print(x)
```

### 求角度
* 根据双曲正弦、余弦、正切值求角度。例如：sinh、cosh 和 tanh 的反函数（arcsinh、arccosh、arctanh）。
* Numpy 提供了 ufunc arcsinh()、arccosh() 和 arctanh()，它们根据给定的相应 sinh、cosh 和 tanh 值生成弧度值。

```python
#  求 1.0 的角度：
x = np.arcsinh(1.0)
print(x)

# 求数组中所有 tanh 值的角度：
arr = np.array([0.1, 0.2, 0.5])
x = np.arctanh(arr)
print(x)
```
